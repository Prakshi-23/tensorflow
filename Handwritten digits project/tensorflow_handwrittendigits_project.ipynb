```python
import tensorflow as tf
from tensorflow import keras
```

```python
from  tensorflow.keras.datasets import mnist
```

```python
# load adat
 (x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(x_train[1])
plt.title(y_train[1])
```

```python
x_train[1]
```

```python
# normalizing the data
# scaling -> values between 0 to 1
#  pixel size 0 to 255
x_train = x_train/255.
x_test = x_test/255.
```

```python
#  reshaping (convo2d - height, wif=dth, channel (RGB BASED)) no channel then black n white
# rest shape is same, 1 is added
x_train = x_train.reshape((x_train.shape[0],x_train.shape[1],x_train.shape[2],1))
x_test = x_test.reshape((x_test.shape[0],x_test.shape[1],x_test.shape[2],1))
```

```python
x_train.shape, x_test.shape
```

```python
y_train[:10]
```

```python
# y was label encoded so we did one-hot encodjing
# same as pd.get_dummies
from tensorflow.keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
# after one-hot encoding
y_test[:10]
```

```python
from keras.layers import Conv2D,Dense,Flatten,MaxPool2D
from keras.models import Sequential
```

```python
x_train.shape
```

```python
model = Sequential()
# Layer
model.add(Conv2D(64,3, activation='relu', input_shape =(28,28,1)))
# Layer any shape but same (3,3)(4,4)
model.add(MaxPool2D((2,2)))
# changing multi dim to single 1D
# converts (28,28,1) in 1D array while sending it to dense layer
model.add(Flatten())
#  number of classes=  number of neurons in dense
model.add(Dense(10,activation='softmax'))
```

```python
print(model.summary())
```

```python
# Compilation
model.compile(optimizer ='adam',loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# number of iterations or whole data rins that many times
h = model.fit(x_train, y_train, epochs=10, validation_data=(x_test,y_test))
```

```python
plt.plot(h.history['accuracy'],label='train acc')
plt.plot(h.history['val_accuracy'],label='test acc')
plt.legend()
```

```python
plt.plot(h.history['loss'],label='train loss')
plt.plot(h.history['val_loss'],label='test loss')
plt.legend()
```
